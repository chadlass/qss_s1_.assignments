```python
import pandas as pd
import numpy as np
from plotnine import *
from plotnine.data import *
```

```python
dol = pd.read_csv("dol_dataset.csv")
dol.head()
dol.columns
```

```python
## add ratio columns for backwages and CMPS
cmp_df = dol.groupby(['st_cd', 'end_year', 'total_employment']).agg({'h2a_cmp_assd_amt': ['mean', 'sum', 'count'],
                                                           'h2a_violtn_cnt' : 'sum'})

cmp_df.columns = cmp_df.columns.droplevel(0)
cmp_df.columns = ['h2a_cmp_amt_mean', 'h2a_cmp_amt_sum', 'cmp_cases',
                  'h2a_violtn_cnt_sum']

cmp_df.reset_index()

##ratio of cases actually resulting in civil monetary penalities
cmp_df['amt_pv'] = cmp_df['h2a_cmp_amt_sum'] / cmp_df['h2a_violtn_cnt_sum']
cmp_df.reset_index()
```

```python
p = (ggplot(aes(x='total_employment', y='amt_pv'), cmp_df))
p + geom_point()
```
